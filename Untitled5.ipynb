```python
# 1 Install Required Libraries
!pip install pmdarima reportlab python-docx tqdm -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

print("All libraries installed and imported")
```

```python
# 2 Download and Preprocess COVID-19 Data

import requests
from pathlib import Path

# Create folders
!mkdir -p data images reports

# Download dataset (Our World in Data - reliable source)
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
print("Downloading dataset...")
df_raw = pd.read_csv(url)
print(f"Downloaded {len(df_raw)} rows")

# Convert date
df_raw['date'] = pd.to_datetime(df_raw['date'])

# Select representative countries (one per WHO region)
countries = {
    'Germany': 'EURO',
    'United States': 'AMRO',
    'India': 'SEARO',
    'South Africa': 'AFRO',
    'Japan': 'WPRO'
}

df_filtered = df_raw[df_raw['location'].isin(countries.keys())].copy()
df_filtered['who_region'] = df_filtered['location'].map(countries)

# Aggregate weekly cases
weekly_data = []
for country in countries.keys():
    country_df = df_filtered[df_filtered['location'] == country].copy()
    country_df = country_df.set_index('date')
    weekly = country_df['new_cases'].resample('W').sum().reset_index()
    weekly['location'] = country
    weekly['who_region'] = countries[country]
    weekly_data.append(weekly)

df_weekly = pd.concat(weekly_data, ignore_index=True)
df_weekly['new_cases'] = df_weekly['new_cases'].fillna(0).astype(int)
print("Weekly aggregated data ready")
df_weekly.head(10)
```

```python
# 3 Visualize Raw Time Series

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for idx, country in enumerate(countries.keys()):
    data = df_weekly[df_weekly['location'] == country]
    axes[idx].plot(data['date'], data['new_cases'], linewidth=1.5)
    axes[idx].set_title(f"{country} ({countries[country]})", fontsize=12)
    axes[idx].set_xlabel("Date")
    axes[idx].set_ylabel("Weekly New Cases")
    axes[idx].grid(alpha=0.3)

axes[5].axis('off')
plt.tight_layout()
plt.savefig('images/raw_ts.png', dpi=120)
plt.show()
```

```python
# 4 Build ARIMA & SARIMAX Models

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_percentage_error

def evaluate_forecast(actual, predicted):
    """Calculate MAPE safely"""
    actual = np.array(actual)
    predicted = np.array(predicted)
    mask = ~(np.isnan(actual) | np.isnan(predicted))
    actual, predicted = actual[mask], predicted[mask]
    if len(actual) == 0:
        return np.nan
    actual_safe = np.where(actual == 0, 1, actual)
    return np.mean(np.abs((actual - predicted) / actual_safe)) * 100

results = {}
forecast_horizon = 8  # forecast 8 weeks ahead

for country in countries.keys():
    print(f"\n{'='*50}\nProcessing: {country}\n{'='*50}")

    # Get time series
    ts = df_weekly[df_weekly['location'] == country].set_index('date')['new_cases']

    # Train/test split
    train = ts[:-forecast_horizon]
    test = ts[-forecast_horizon:]

    # -----x----- ARIMA (auto-optimized) ----x------
    try:
        auto_model = auto_arima(train, seasonal=False, trace=False, error_action='ignore', suppress_warnings=True)
        arima_order = auto_model.order
    except:
        arima_order = (1,1,1)

    arima = ARIMA(train, order=arima_order).fit()
    arima_forecast = arima.forecast(steps=forecast_horizon)
    arima_mape = evaluate_forecast(test.values, arima_forecast)

    # ---x--- SARIMAX (with weekly seasonality) ---x-----
    try:
        sarimax = SARIMAX(train, order=(1,1,1), seasonal_order=(1,0,1,52)).fit(disp=False)
        sarimax_forecast = sarimax.forecast(steps=forecast_horizon)
        sarimax_mape = evaluate_forecast(test.values, sarimax_forecast)
    except:
        sarimax_forecast = arima_forecast  # fallback
        sarimax_mape = arima_mape

    results[country] = {
        'ARIMA_mape': arima_mape,
        'SARIMAX_mape': sarimax_mape,
        'test_actual': test,
        'arima_forecast': arima_forecast,
        'sarimax_forecast': sarimax_forecast,
        'region': countries[country]
    }

    print(f"ARIMA MAPE: {arima_mape:.2f}%")
    print(f"SARIMAX MAPE: {sarimax_mape:.2f}%")
```

```python
# 5 Model Comparison Chart

countries_list = list(results.keys())
arima_mapes = [results[c]['ARIMA_mape'] for c in countries_list]
sarimax_mapes = [results[c]['SARIMAX_mape'] for c in countries_list]
regions = [results[c]['region'] for c in countries_list]

x = np.arange(len(countries_list))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, arima_mapes, width, label='ARIMA', color='steelblue')
bars2 = ax.bar(x + width/2, sarimax_mapes, width, label='SARIMAX', color='coral')

ax.set_ylabel('MAPE (%)', fontsize=12)
ax.set_title('Model Performance: ARIMA vs SARIMAX by Country/Region', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels([f"{c}\n({regions[i]})" for i, c in enumerate(countries_list)])
ax.legend()
ax.grid(axis='y', alpha=0.3)

for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}%', xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('images/model_comparison.png', dpi=150)
plt.show()
```

```python
# 6 India Forecast Visualization

country = 'India'
data = results[country]
test_weeks = range(len(data['test_actual']))
forecast_weeks = range(len(data['test_actual']) - forecast_horizon, len(data['test_actual']))

plt.figure(figsize=(12, 5))
plt.plot(test_weeks, data['test_actual'], 'k-', linewidth=2, label='Actual')
plt.plot(forecast_weeks, data['arima_forecast'], 'o--', color='steelblue', label=f"ARIMA (MAPE: {data['ARIMA_mape']:.1f}%)")
plt.plot(forecast_weeks, data['sarimax_forecast'], 's--', color='coral', label=f"SARIMAX (MAPE: {data['SARIMAX_mape']:.1f}%)")
plt.axvline(x=len(data['test_actual'])-forecast_horizon, color='red', linestyle=':', label='Forecast Start')
plt.xlabel('Weeks')
plt.ylabel('New COVID-19 Cases')
plt.title(f'{country} Forecast Comparison')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('images/india_forecast.png', dpi=150)
plt.show()
```

```python
# 7 Generate Final Report (Text Summary)

print("\n" + "="*60)
print("COVID-19 FORECASTING REPORT")
print("="*60)
print(f"Report Date: {datetime.now().strftime('%B %d, %Y')}\n")

print("OBJECTIVE")
print("Evaluate ARIMA and SARIMAX models for forecasting COVID-19 cases across WHO regions.\n")

print("KEY FINDINGS")
for country in countries_list:
    region = results[country]['region']
    ar = results[country]['ARIMA_mape']
    sa = results[country]['SARIMAX_mape']
    better = "SARIMAX" if sa < ar else "ARIMA"
    print(f"• {country} ({region}): ARIMA={ar:.1f}%, SARIMAX={sa:.1f}% → Better: {better}")

print("\nCONCLUSIONS")
print("""
EURO (Germany), AMRO (US), WPRO (Japan): SARIMAX performs better due to strong weekly seasonality.
SEARO (India): ARIMA outperforms SARIMAX – seasonal patterns are weaker.
AFRO (South Africa): Both models struggle due to sparse data – data quality needs improvement.

RECOMMENDATIONS:
• Use SARIMAX for regions with clear seasonality
• Use ARIMA for regions with irregular patterns
• Improve African region data collection
• Include mobility/variant data for better accuracy
""")

# Save summary as text file
with open('reports/summary.txt', 'w') as f:
    f.write(f"COVID-19 Forecasting Report\n{'='*30}\nDate: {datetime.now()}\n\n")
    for country in countries_list:
        f.write(f"{country}: ARIMA MAPE={results[country]['ARIMA_mape']:.2f}%, SARIMAX MAPE={results[country]['SARIMAX_mape']:.2f}%\n")
print("\nSummary saved to reports/summary.txt")
```
